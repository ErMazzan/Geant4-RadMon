```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit
```

```python
def read_file(filename, column):
    with open(filename, 'r') as data:
        x = []
        for line in data:
            p = line.split(";")
            x.append(float(p[column]))
    return x
```

```python
def gauss(x, mu, sigma, C):
    return C*np.exp(-(x - mu)**2/(2*sigma**2))
```

```python
energy_names = ["100-MeV","200-MeV","300-MeV","500-MeV","700-MeV","1-GeV","2-GeV"]

energies = []
values = []

fig1, ax1 = plt.subplots(1, 1)

plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)

for name in energy_names:
    filename = "EScint1/{}.txt".format(name)
    energy = read_file(filename, 0)
    counts = read_file(filename, 1)
    
    energies.append(energy)
    values.append(counts)
    
    ax1.set_title("Scintillator 1 - Beam", fontsize=16)
    ax1.set_xlabel("Energy (MeV)", fontsize=14)
    ax1.set_ylabel("Counts", fontsize=14)
    ax1.plot(energies, values, color = "firebrick")
    
    fig1.set_size_inches(12, 6.5)
    
means_scin_1 = []
for i in range(len(energies)):
    energy_fit = []
    values_fit = []
    
    j = 0
    if sum(values[i]) >= 1000:
        for count in values[i]:
            j += 1
            if count >= 500:
                energy_fit.append(energies[i][j])
                values_fit.append(values[i][j])

        popt, pcov = curve_fit(gauss, energy_fit, values_fit)
        means_scin_1.append(popt[0])
        print("Energy:", energy_names[i], " Mean:", popt[0])

ax1.plot(2.995, 1125, marker="o", markersize=10, markerfacecolor="steelblue", markeredgecolor="none", label = "Beam energy")
ax1.text(2.995 + 0.1, 1100, "100 MeV", fontsize = 14)

ax1.plot(1.3905, 1350, marker="o", markersize=10, markerfacecolor="steelblue", markeredgecolor="none")
ax1.text(1.3905 + 0.1, 1325, "200 MeV", fontsize = 14)

ax1.plot(1.0161, 1650, marker="o", markersize=10, markerfacecolor="steelblue", markeredgecolor="none")
ax1.text(1.0161 + 0.1, 1625, "300 MeV", fontsize = 14)
        
ax1.plot(0.7465, 2000, marker="o", markersize=10, markerfacecolor="steelblue", markeredgecolor="none")
ax1.text(0.7465 + 0.1, 1975, "500 MeV", fontsize = 14)

ax1.plot(0.6422, 2250, marker="o", markersize=10, markerfacecolor="steelblue", markeredgecolor="none")
ax1.text(0.6422 + 0.1, 2225, "700 MeV", fontsize = 14)

ax1.plot(0.5680, 2500, marker="o", markersize=10, markerfacecolor="steelblue", markeredgecolor="none")
ax1.text(0.5680 + 0.1, 2475, "1 GeV", fontsize = 14)

ax1.plot(0.4908, 2750, marker="o", markersize=10, markerfacecolor="steelblue", markeredgecolor="none")
ax1.text(0.4908 + 0.1, 2725, "2 GeV", fontsize = 14)

ax1.legend(loc="upper right", prop={'size': 14})
fig1.savefig('figures/Scintillator_1.pdf')
plt.show()
```
