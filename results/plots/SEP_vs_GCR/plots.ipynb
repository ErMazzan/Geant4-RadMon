```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Function to rebin the data. 
# INPUT: bins_div (# of new bins), bins (original bin data), counts (values for each original bin), threshold (maximum
#        energy value, x axis)
# OUTPUT: rebinned_bins (new bins), rebinned_counts (new counts for each new bin created)

def rebin(bins_div, bins, counts, threshold):
    rebin_bins = []
    rebin_counts = []

    for bini, count in zip(bins, counts):
        if bini <= threshold:
            rebin_bins.append(bini)
            rebin_counts.append(count)

    rebin_bins = np.array(rebin_bins)
    rebin_bins = rebin_bins.reshape(bins_div, int(len(rebin_bins)/bins_div))

    rebin_counts = np.array(rebin_counts)
    rebin_counts = rebin_counts.reshape(bins_div, int(len(rebin_counts)/bins_div))

    rebinned_counts = np.sum(rebin_counts, axis=1)

    rebinned_bins = []
    for i in range(len(rebin_bins)): 
        rebinned_bins.append(rebin_bins[i][-1])
        
    return rebinned_bins, rebinned_counts
```

```python
sep_name = "data/sep_EDep2_Scint1&2.txt"
proton_GCR_name = "data/proton_GCR_EDep2_Scint1&2.txt"

bins = np.array(pd.read_csv("data/proton_GCR_EDep2_Scint1&2.txt", header=None, sep=";")[0])

sep_counts = np.array(pd.read_csv(sep_name, header=None, sep=";")[1])
proton_GCR_counts = np.array(pd.read_csv(proton_GCR_name, header=None, sep=";")[1])

threshold = 10
bins_div = 100
bins, sep_counts = rebin(bins_div, bins, sep_counts, threshold)

threshold = 10
bins_div = 100
bins, proton_GCR_counts = rebin(bins_div, bins, proton_GCR_counts, threshold)

fig, ax = plt.subplots(1,1)

ax.set_xlim([0, threshold])

ax.set_xlabel("Energy (MeV)", fontsize=18)
ax.set_ylabel("Counts", fontsize=18)

ax.tick_params(axis='both', which='major', direction = "out", labelsize=18, size = 12)

ax.step(bins, sep_counts, linestyle = "-", linewidth = 1.5,
        color = "firebrick", label = "SEP")

ax.step(bins, proton_GCR_counts, linestyle = "-", linewidth = 1.5,
        color = "blue", label = "GCR")

ax.set_title("Deposited energy in Scintillator 2 - Coinc 1-2", fontsize = 14)

ax.legend(loc="upper right", fontsize = 14)

fig.set_size_inches(10.5, 4.5)
plt.show()
```

**Important note**: Results above ARE NOT TIME SCALED, they are only the results extracted from the simulations.

## Time scaled results

```python
# Function to time scale the data obtained from the simulations
# INPUT: data (original non-time scaled data), sim_particles (# of total initial particles simulated), 
#        flux (in units of #part./(s cm^2)), r (radius of the simulated sphere, in cm), time (time scaling of the data,
#        given in seconds)) 
# OUTPUT: data (time scaled data)

def time_scale(data, sim_particles, flux, r, time):
    time_scaling = sim_particles/(flux*4*np.pi*r*r) # Units are set to seconds 
    time_scaling = time/time_scaling
    data = data*time_scaling
    
    return data
```

```python
sep_name = "data/sep_EDep2_Scint1&2.txt"
proton_GCR_name = "data/proton_GCR_EDep2_Scint1&2.txt"

bins = np.array(pd.read_csv("data/proton_GCR_EDep2_Scint1&2.txt", header=None, sep=";")[0])

sep_counts = np.array(pd.read_csv(sep_name, header=None, sep=";")[1])
proton_GCR_counts = np.array(pd.read_csv(proton_GCR_name, header=None, sep=";")[1])

threshold = 10
bins_div = 100
bins, sep_counts = rebin(bins_div, bins, sep_counts, threshold)

threshold = 10
bins_div = 100
bins, proton_GCR_counts = rebin(bins_div, bins, proton_GCR_counts, threshold)

sim_particles = 5e6
flux = 1.89
r = 7
time = 3600*24*365 # We set to 1 year
proton_GCR_counts = time_scale(proton_GCR_counts, sim_particles, flux, r, time)

sim_particles = 1e8
flux = 5.634e7
r = 7
time = 60 # We set to 1 minute
sep_counts = time_scale(sep_counts, sim_particles, flux, r, time)
```

```python
fig, axes = plt.subplots(1,1)

axes.set_xlim([0, threshold])

axes.set_xlabel("Deposited energy (MeV)", fontsize=18)
axes.set_ylabel("Counts", fontsize=18)

axes.tick_params(axis='both', which='major', direction = "out", labelsize=18, size = 12)
axes.tick_params(axis='both', which='minor', direction = "out", labelsize=18, size = 6)

axes.step(bins, sep_counts, linestyle = "-", linewidth = 1.5,
        color = "firebrick", label = "SEP (20 May 2001) ~ 1 minute")

axes.step(bins, proton_GCR_counts, linestyle = "-", linewidth = 1.5,
        color = "blue", label = "GCR protons solar maximum ~ 1 year")

axes.set_yscale("log")

axes.legend(loc="upper right", fontsize = 14)
    
axes.set_title('Scintillator 2 - Coincidence 1 & 2', fontsize = 18)

fig.tight_layout()
fig.set_size_inches(10.5, 8.5)
fig.savefig("plots/proton_GCR_vs_SEP_EDep2_Coinc1-2.pdf")
fig.savefig("plots/proton_GCR_vs_SEP_EDep2_Coinc1-2.jpg")
plt.show()
```
