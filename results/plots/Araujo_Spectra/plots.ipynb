```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
p_solarmin_name = "data/p_solarmin.csv"
p_solarmax_name = "data/p_solarmax.csv"  
he_solarmax_name = "data/he_4_solarmax.csv"
he_solarmin_name = "data/he_4_solarmin.csv"
sep_name = "data/sep.csv"

p_solarmin_energy = np.array(pd.read_csv(p_solarmin_name, header=None, sep=",")[0])*1000
p_solarmin_flux = np.array(pd.read_csv(p_solarmin_name, header=None, sep=",")[1])

p_solarmax_energy = np.array(pd.read_csv(p_solarmax_name, header=None, sep=",")[0])*1000
p_solarmax_flux = np.array(pd.read_csv(p_solarmax_name, header=None, sep=",")[1])

he_solarmax_energy = np.array(pd.read_csv(he_solarmax_name, header=None, sep=",")[0])*1000
he_solarmax_flux = np.array(pd.read_csv(he_solarmax_name, header=None, sep=",")[1])

he_solarmin_energy = np.array(pd.read_csv(he_solarmin_name, header=None, sep=",")[0])*1000
he_solarmin_flux = np.array(pd.read_csv(he_solarmin_name, header=None, sep=",")[1])

sep_energy = np.array(pd.read_csv(sep_name, header=None, sep=",")[0])*1000
sep_flux = np.array(pd.read_csv(sep_name, header=None, sep=",")[1])*10000
```

```python
energies = [p_solarmax_energy, p_solarmin_energy, he_solarmax_energy, he_solarmin_energy]
fluxes = [p_solarmax_flux, p_solarmin_flux, he_solarmax_flux, he_solarmin_flux]
colors = ["firebrick", "blue", "green", "purple", "black", "orange", "darkblue"]
labels = ["p - solarmax", "p - solarmin", "He-4 - solarmax", "He-4 - solarmin"]

fig, axes = plt.subplots(1,2)

##################################
##### PLOTS PROTONS & HELIUM #####
##################################

axes[1].set_xlabel("Energy (MeV)", fontsize=14)
axes[1].set_ylabel(r"Flux $\left ( \frac{particles}{s \cdot m^2 \cdot sr \cdot MeV} \right )$",fontsize=14)

axes[1].tick_params(axis='both', which='major', direction = "out", labelsize=14, size = 12)
axes[1].tick_params(axis='both', which='minor', direction = "out", labelsize=14, size = 6)

for i in range(len(energies)):
    axes[1].plot(energies[i], fluxes[i], "-", color = colors[i], 
             label = labels[i])

axes[1].set_xscale("log")
axes[1].set_yscale("log")

#axes[1].set_xlim([0.1, 400])

axes[1].legend(loc="upper right", fontsize = 10)

###########################
##### PLOTS SEP & GCR #####
###########################

axes[0].set_xlabel("Energy (MeV)", fontsize=14)
axes[0].set_ylabel(r"Flux $\left ( \frac{particles}{s \cdot m^2 \cdot sr \cdot MeV} \right )$",fontsize=14)

axes[0].tick_params(axis='both', which='major', direction = "out", labelsize=14, size = 12)
axes[0].tick_params(axis='both', which='minor', direction = "out", labelsize=14, size = 6)

axes[0].plot(sep_energy, sep_flux, "-", color = "firebrick", label = "SEP")
axes[0].plot(p_solarmax_energy, p_solarmax_flux, "-", color = "blue", label = "p - solarmax")
axes[0].plot(p_solarmin_energy, p_solarmin_flux, "-", color = "green", label = "p - solarmin")

axes[0].set_xscale("log")
axes[0].set_yscale("log")

axes[0].legend(loc="upper right", fontsize = 10)

fig.set_size_inches(10.5, 6.5)
fig.tight_layout()
plt.show()
```
